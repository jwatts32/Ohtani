<h2>Import libraries

```python
import pandas as pd
```

<h2>Get data for LA Angels for last 3 seasons, pass each season to dataframe, and get 1st table from page

```python
url_laa_2023 = "https://www.baseball-reference.com/teams/LAA/2023.shtml"
url_laa_2022 = "https://www.baseball-reference.com/teams/LAA/2022.shtml"
url_laa_2021 = "https://www.baseball-reference.com/teams/LAA/2021.shtml"

laa_2023 = pd.read_html(url_laa_2023)
laa_2022 = pd.read_html(url_laa_2022)
laa_2021 = pd.read_html(url_laa_2021)

laa_2023 = laa_2023[0]
laa_2022 = laa_2022[0]
laa_2021 = laa_2021[0]
```

<h2>Add "Season" field to each dataframe for each respective season.

```python
laa_2023["Season"] = 2023
laa_2022["Season"] = 2022
laa_2021["Season"] = 2021
```

<h2>Append dataframes to create a single dataframe, display data sorted by name to valide duplicate names are present since there are multiple seasons

```python
angels = laa_2023.append(laa_2022).append(laa_2021)
display(angels.sort_values("Name"))
```

<h2>Query for non-pitchers only

```python
angels = angels.query('Pos not in ("P","Pos") ')
display(angels.sort_values("Name"))
```

<h2>Drop rows that have missing values.

```python
angels = angels.dropna()
display(angels.sort_values("Name").head(20))
```

<h2>Subset data for required fields.

```python
angels = angels[["Season","Name","G","PA","AB","H","BA","HR","RBI","R"]]
display(angels.sort_values("Name").head(50))
```

<h2>Check data types to see if they are already in correct state for calculating.

```python
angels.dtypes
```

<h2>Change data types for appropriate fields.

```python
angels = angels.astype({"G": "int", "PA": "int", "AB": "int", "H": "int", "BA": "float", "HR": "int", "RBI": "int", "R": "int"})
angels.dtypes
```

<h2>Query for players who have at least 250 plate appearances.

```python
angels = angels.query('PA >= 250')
display(angels.sort_values("Name"))
```

<h2>Get averages for Ohtani only.

```python
avgs_sho_only = pd.DataFrame(angels.query('Name == "Shohei Ohtani*" ').agg({"H": "mean","BA": "mean", "HR": "mean", "RBI": "mean", "R": "mean"}))
display(avgs_sho_only)
```

<h2>Give column a name.

```python
avgs_sho_only.columns = ["Ohtani"]
display(avgs_sho_only)
```

<h2>Get team averages, excluding Ohtani

```python
team_avgs_no_sho = pd.DataFrame(angels.query('Name != "Shohei Ohtani*" ').agg({"H": "mean","BA": "mean", "HR": "mean", "RBI": "mean", "R": "mean"}))
display(team_avgs_no_sho)
```

<h2>Give column a name.

```python
team_avgs_no_sho.columns = ["Rest of Team"]
display(team_avgs_no_sho)
```

<h2>Get team averages with Ohtani included.

```python
team_avgs_with_sho = pd.DataFrame(angels.agg({"H": "mean","BA": "mean", "HR": "mean", "RBI": "mean", "R": "mean"}))
display(team_avgs_with_sho)
```

<h2>Give column a name.

```python
team_avgs_with_sho.columns = ["Team with Ohtani"]
display(team_avgs_with_sho)
```

<h2>Concatenate the 3 dataframes into a single dataframe.

```python
angels_final_df = pd.concat([avgs_sho_only, team_avgs_no_sho, team_avgs_with_sho], axis="columns")
display(angels_final_df)
```

<h2>Round field rows to appropriate number of decimal places.

```python
angels_final_df.iloc[0] = round(angels_final_df.iloc[0],0)
angels_final_df.iloc[1] = round(angels_final_df.iloc[1],3)
angels_final_df.iloc[2] = round(angels_final_df.iloc[2],0)
angels_final_df.iloc[3] = round(angels_final_df.iloc[3],0)
angels_final_df.iloc[4] = round(angels_final_df.iloc[4],0)
display(angels_final_df)
```

<h2>I had issues changing the data types to integers at the row level, so I flipped the dataframe using transpose().

```python
angels_final_df = angels_final_df.transpose()
display(angels_final_df)
```

<h2>Now, I can convert the data types by column.

```python
angels_final_df = angels_final_df.astype({"H": "int", "HR": "int", "RBI": "int", "R": "int"})
display(angels_final_df)
```
